```python
import cv2
import numpy as np


image = cv2.imread('/Users/dishanimandal/Desktop/dip/image.png')  # Replace with your image path

height, width = image.shape[:2]
resized_image = cv2.resize(image, (width // 4, height // 4))


sharpen_kernel = np.array([[ 0, -1,  0],
                           [-1,  5, -1],
                           [ 0, -1,  0]])


sharpened = cv2.filter2D(resized_image, -1, sharpen_kernel)


cv2.imshow("Original (Resized)", resized_image)
cv2.imshow("Sharpened (Resized)", sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img = cv2.imread('MaskImage.jpg')
cv2.imshow('Original Image', img)
HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_yellow = np.array([15,50,180])
upper_yellow = np.array([40,255,255])
mask = cv2.inRange(HSV, lower_yellow, upper_yellow)
cv2.imshow('Mask', mask)
result = cv2.bitwise_and(img, img, mask=mask)
cv2.imshow('masked image', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('sample.png')  
img = cv2.resize(img, (500, 400)) 
def nothing(x):
    pass  # Empty function required by createTrackbar

cv2.namedWindow("Brightness Adjuster",cv2.WINDOW_NORMAL)

cv2.createTrackbar('Brightness', 'Brightness Adjuster', 50, 100, nothing)

while True:
    
    brightness = cv2.getTrackbarPos('Brightness', 'Brightness Adjuster')
    bright_img = cv2.convertScaleAbs(img, alpha=1, beta=brightness - 50)

    cv2.imshow('Brightness Adjuster', bright_img)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def onChange(val):
    pass

def createTrackBar():
    cv2.namedWindow("mywindow")
    cv2.resizeWindow("mywindow", 512, 150)
    cv2.createTrackbar("red", "mywindow", 0, 255, onChange)
    cv2.createTrackbar("green", "mywindow", 0, 255, onChange)
    cv2.createTrackbar("blue", "mywindow", 0, 255, onChange)

createTrackBar()

image = np.zeros((512, 512, 3), np.uint8)

while True:
    # Get trackbar positions
    red = cv2.getTrackbarPos("red", "mywindow")
    green = cv2.getTrackbarPos("green", "mywindow")
    blue = cv2.getTrackbarPos("blue", "mywindow")

    image[:] = [blue, green, red]
    cv2.imshow("mywindow", image)

    # Exit on Esc key
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
